```python
length = 45
width = 35
#Koordinat foto
x = 0
y = 0
type(Person3)
```

```python
Conc = np.concatenate((Person1, Person2, Person3, Person4, Person5, Person6, Person7, Person8, Person9, Person10))
Conc.shape
plt.imshow(Conc, cmap='gray', vmin=0, vmax=255)
```

```python
Conc.shape
```

```python
Train_Raw = np.zeros([Conc.shape[0], 
                      round(Conc.shape[1]/width/2+0.1)*width])
Val_Raw = np.zeros([Conc.shape[0], 
                   (37-round(Conc.shape[1]/width/2+0.1))*width])
print(Train_Raw.shape)
print(Val_Raw.shape)
```

```python
y=0
for i in range(0,19):
    Train_Raw[:,width*i:width+width*i] = Conc[:,width*y:width+width*y]
    y=y+2
y=1
for i in range(0,18):
    Val_Raw[:,width*i:width+width*i] = Conc[:,width*y:width+width*y]
    y=y+2
print(Train_Raw.shape)
print(Val_Raw.shape)
```

```python
plt.imshow(Train_Raw, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imshow(Val_Raw, cmap='gray', vmin=0, vmax=255)
```
