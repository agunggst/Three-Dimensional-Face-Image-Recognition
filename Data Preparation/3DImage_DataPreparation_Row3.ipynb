```python
def Val_gen(r,Person):
    i=10
    Val_Raw[length*(0+i*r):length+length*(0+i*r),:] = Person[length*0:length+length*0,:]
    Val_Raw[length*(1+i*r):length+length*(1+i*r),:] = Person[length*2:length+length*2,:]
    Val_Raw[length*(2+i*r):length+length*(2+i*r),:] = Person[length*3:length+length*3,:]
    Val_Raw[length*(3+i*r):length+length*(3+i*r),:] = Person[length*5:length+length*5,:]
    Val_Raw[length*(4+i*r):length+length*(4+i*r),:] = Person[length*6:length+length*6,:]
    Val_Raw[length*(5+i*r):length+length*(5+i*r),:] = Person[length*8:length+length*8,:]
    Val_Raw[length*(6+i*r):length+length*(6+i*r),:] = Person[length*9:length+length*9,:]
    Val_Raw[length*(7+i*r):length+length*(7+i*r),:] = Person[length*11:length+length*11,:]
    Val_Raw[length*(8+i*r):length+length*(8+i*r),:] = Person[length*12:length+length*12,:]
    Val_Raw[length*(9+i*r):length+length*(9+i*r),:] = Person[length*14:length+length*14,:]
    return
```

```python
Val_gen(0,Person1)
Val_gen(1,Person2)
Val_gen(2,Person3)
Val_gen(3,Person4)
Val_gen(4,Person5)
Val_gen(5,Person6)
Val_gen(6,Person7)
Val_gen(7,Person8)
Val_gen(8,Person9)
Val_gen(9,Person10)
```
